```python
import pandas as pd
```

## Read

```python
portfolio = pd.read_json('../data/raw/portfolio.json', orient='records', lines=True)
portfolio.head()
```

## Rename

```python
portfolio = portfolio.rename(columns={'id': 'portfolio_id'})
portfolio.head()
```

## Clean

```python
portfolio = portfolio.explode('channels')
portfolio.head()
```

## Save

```python
portfolio.to_csv('../data/processed/portfolio.csv', index=False)
```
