```python
import pandas as pd
```

## Read

```python
portfolio = pd.read_csv('../data/processed/portfolio.csv')
portfolio.head()
```

```python
profile = pd.read_csv('../data/processed/profile.csv')
profile.head()
```

```python
transcript = pd.read_csv('../data/processed/transcript.csv')
transcript.head()
```

## Join

```python
user_behaviour = pd.merge(transcript ,portfolio, on="portfolio_id", how="left")
user_behaviour = pd.merge(user_behaviour,profile, on="profile_id", how="left")
user_behaviour.head()
```

## Save

```python
user_behaviour.to_csv('../data/processed/user_behaviour.csv', index=False)
```
