```python
import pandas as pd
```

## Load

```python
portfolio = pd.read_csv('../../data/features/portfolio.csv')
portfolio.head()
```

```python
profile = pd.read_csv('../../data/features/profile.csv')
profile.head()
```

```python
transcript = pd.read_csv('../../data/features/transcript.csv')
transcript.head()
```

## Add features

```python
user_item = pd.merge(transcript ,portfolio[["portfolio_id","offer"]], on="portfolio_id", how="left")
user_item = pd.merge(user_item,profile[["profile_id","customer_type"]], on="profile_id", how="left")
user_item[["profile_id","offer","score","customer_type"]].head()
```

## Save

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(user_item[["profile_id","offer","score","customer_type"]], test_size=0.2, random_state=42)

print("Train set:")
print(train.info())
print("\nTest set:")
print(test.info())
```

```python
train.to_csv('../../data/train/user_item.csv', index=False)
```

```python
test.to_csv('../../data/test/user_item.csv', index=False)
```
